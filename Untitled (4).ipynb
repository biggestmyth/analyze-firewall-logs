creating the data frame:

```python
import pandas as pd 
import requests
import matplotlib.pyplot as plt
import urllib

firewall_logs=pd.read_csv("firewall logs.csv")
firewall_logs.head(10)
```

analyze actions


```python
action_counts=firewall_logs['Action'].value_counts()
print("action counts:")
print(action_counts)
```

making blocked ports

```python
targeted_ports=firewall_logs[firewall_logs['Action']=='Blocked']['Port'].value_counts()
top_targeted_ports=targeted_ports[targeted_ports > 2]
print(top_targeted_ports)
```

making porotocol distributuin

```python
porotocol_distributuin=firewall_logs['Protocol'].value_counts()
print(porotocol_distributuin)
```

unauthorized access on specific port

```python

unauthorized_access=firewall_logs [( firewall_logs ['Action']=='Blocked') & (firewall_logs['Port']=='22')]['Source IP'].value_counts()
print (unauthorized_access)
```

minute blocked trends

```python
firewall_logs["DateTime"]=firewall_logs["DateTime"].apply(pd.to_datetime)
time_series=firewall_logs[firewall_logs['Action']=='Blocked'].groupby(firewall_logs["DateTime"].dt.minute).size()
```

count actions by porotocol

```python
actions_by_protocol=firewall_logs.groupby('Protocol')['Action'].value_counts().unstack().fillna(0)
print("\nActions by Protocol:")
print(actions_by_protocol)
```

count action by rule

```python
actions_by_rule=firewall_logs.groupby('Rule')['Action'].value_counts().unstack().fillna(0)
print("\nActions by Rule:")
print(actions_by_rule)
```

analyze protocol distribution for blocked and allowed ips

```python
blocked_ips=firewall_logs[firewall_logs['Action'] == 'Blocked']
allowed_ips=firewall_logs[firewall_logs['Action'] == 'Allowed']

# Analyze protocol distribution for blocked and allowed IPs
protocol_distribution_blocked=blocked_ips['Protocol'].value_counts()
protocol_distribution_allowed=allowed_ips['Protocol'].value_counts()
```

top rules for blocking actions


```python
top_blocking_rules=firewall_logs[firewall_logs['Action']=='Blocked']['Rule'].value_counts().head(10)
print("\nTop Rules for Blocking Actions:")
print(top_blocking_rules)
```

Plot action distribution

```python
plt.figure(figsize=(10, 6))
action_counts.plot(kind='bar', color=['green', 'red'])
plt.title('Distribution of Actions (Allowed vs. Blocked)')
plt.xlabel('Action')
plt.ylabel('Count')
plt.show()
```

protocol distrbution plot

```python
porotocol_distributuin.plot(kind='pie',autopct ='%1.1f%%', startangle=90)
plt.title("protocol distribuyion")
plt.ylabel("")
plt.show()
```

top targeted ports plot

```python
top_targeted_ports.plot(kind='bar',color='blue')
plt.title("most targeted ports")
plt.xlabel("port")
plt.ylabel("block count")
```

minute blocked trends plot

```python

time_series.plot(kind='area')
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

actions by protocol plot

```python
actions_by_protocol.plot(kind='bar',stacked=True, figsize=(12, 6))
plt.title('Actions by Protocol')
plt.xlabel('Protocol')
plt.ylabel('Count')
plt.show()
```

actions by rule plot


```python

actions_by_rule.plot(kind='bar', stacked=True,figsize=(12, 6))
plt.title('Actions by Rule')
plt.xlabel('Rule')
plt.ylabel('Count')
plt.show()
```

protocl distrbution for blocked and allowed ips

```python

plt.figure(figsize=(12, 6))
protocol_distribution_blocked.plot(kind='bar', alpha=0.7, label='Blocked', color='red')
protocol_distribution_allowed.plot(kind='bar', alpha=0.7, label='Allowed', color='green')
plt.title('Protocol Distribution (Blocked vs. Allowed)')
plt.xlabel('Protocol')
plt.ylabel('Count')
plt.legend()
plt.show()
```

top rules for blocking actions plot 


```python

plt.figure(figsize=(12, 6))
top_blocking_rules.plot(kind='bar', color='orange')
plt.title('Top Rules for Blocking Actions')
plt.xlabel('Rule')
plt.ylabel('Count')
plt.show()
```
